###### **Creating Graph** (First get all the columns of sources then count how many yes there are for each column; based on that number create a graph and display the top 3 results)
Source: https://stackoverflow.com/questions/57967761/how-to-sum-up-yes-and-no-into-a-total-quantity-using-matplotlib-pandas-python
https://www.w3schools.com/python/matplotlib_bars.asp
ChatGPT Landa

```python
# get columns
sourceColumns = [x for x in df24.columns if x.startswith('hear') or x.startswith('source')]

print("Columns:")
print(sourceColumns)

# count number of columns with a yes
sourceCounts = {}
for col in sourceColumns:
    count = 0
    for x in df24[col]:
        if str(x) == 'Yes':
            count += 1
    sourceCounts[col] = count

# CHATGPT FOR TOP 3 DICTIONARY (LAMBDA) sorting the diction and replacnig for easier plot
sortedSources = sorted(sourceCounts.items(), key=lambda item: item[1], reverse=True)

print("Columns:")
print(sortedSources)

# get top 3
sortedSources = dict(sortedSources[:3])

print("3 Columns:")
print(sortedSources)

# graph
plt.bar(sortedSources.keys(), sortedSources.values())
plt.ylabel('Number of Students')
plt.xlabel('Sources')
plt.title('Top Most Effective Sources for Student Awareness of CCM')
plt.show()
```

##### What sources are most effective in making students aware of CCM?:

##### Family Friends, Highschool Counselors, and the CCM Website are the top sources inorder to make students aware of CCM.

###### **Creating Graph** (First get all the columns for d20 and d24 for the interest level then combine both dfs into a singular one and plot) use percentage to account for voting differences
Source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html
https://www.w3schools.com/python/pandas/pandas_dataframes.asp
https://www.w3schools.com/python/matplotlib_bars.asp

```python
# get data 2020 and 2024 and combine values of interest level to vs
# Define interest levels in order
levels = [1, 2, 3, 4, 5]

# Get counts and reindex so both years align
df = pd.DataFrame({
    '2020': df20['interest_level'].value_counts(normalize=True).reindex(levels, fill_value=0) * 100,
    '2024': df24['interest_level'].value_counts(normalize=True).reindex(levels, fill_value=0) * 100
})

print("temp graph")
print(df)

# graph both based on interest
df.plot(kind='bar')
plt.xlabel('Interest Level (1 = Low, 5 = High)')
plt.ylabel('Percentage of Students')
plt.title('Interest in Taking More Computing Classes (2020 vs 2024)')
plt.legend(title='Year')
plt.show()
```

##### How did the interest level (1-5) of taking more computing classes differ from the year 2020 and 2024?

##### It seems like theres an increase in overall interest in computing classes. As of 2024, about 40% of students were at level 4 and were interested in joining computing related classes wheras in 2020 the majority (38%) were at level 3 interested in joining. 

###### **Creating Graph** (First get all the columns of impact then count how many high impact there are for each column; based on that number create a graph and display the top 5 results) (vs graph comparing 2020 vs 2024); use percentage to account for voting differences
sources:
https://stackoverflow.com/questions/19073683/how-to-fix-overlapping-annotations-text
ChatGPT Landa

```python
# get columns
impactCol = [c for c in df24.columns if c.startswith('impact')]

#empty dict
impact24 = {}
impact20 = {}

# get the sums of high impact in both of the df

total_24 = len(df24)
total_20 = len(df20)


for x in impactCol:
    impact24[x] = (df24[x] == 'High Impact').sum() / total_24 * 100
    
    if x in df20.columns:
        impact20[x] = (df20[x] == 'High Impact').sum() / total_20 * 100

print("\n2024 High Impact percentages:")
print(impact24)
print("\n2020 High Impact percentages:")
print(impact20)

# LAMBDA FROM CHATGPT FROM THE PREVIOUS QUESTION
top24 = dict(sorted(impact24.items(), key=lambda x: x[1], reverse=True)[:5])
top20 = {k: impact20.get(k, 0) for k in top24.keys()}

# Help from CHATGPT To fix the Overlap issue with graphs going ontop of eachother (([i - 0.2 for i in x])
x = range(len(top24))

print("top 24")
print(top24)

print("top 20")
print(top20)

#plot
plt.barh([i - 0.2 for i in x], top20.values(), height=0.4, label='2020')
plt.barh([i + 0.2 for i in x], top24.values(), height=0.4, label='2024')

plt.yticks(x, top24.keys(), rotation=15)
plt.xlabel('Percentage of Students')
plt.ylabel('Impact Factor')
plt.title('Top 5 Enrollment Factors (2020 vs 2024)')
plt.legend(title='Year')
plt.show()
```

##### What are the top factors in students' decisions for enrollment (2020 vs 2024)?

##### It seems like the overall most effective enrollment factors has changed slightly from 2024 and 2020. Back in 2020, the most impactful factor was location. However, in 4 years that changed to transferability, with location now being 3rd in the impact amount.
